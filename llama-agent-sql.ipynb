```python
# Let's add an agent with a tool

# Let's start with a subquestion generator
from llama_index.tools import QueryEngineTool, ToolMetadata
from llama_index.query_engine import SubQuestionQueryEngine
from llama_index.callbacks import CallbackManager, LlamaDebugHandler
from llama_index import SQLDatabase, ServiceContext
from llama_index.llms import OpenAI
```

```python
# Using the LlamaDebugHandler to print the trace of the sub questions
# captured by the SUB_QUESTION callback event type

#llm = OpenAI(model="gpt-4-1106-preview")
llm = OpenAI(model="gpt-4")
llama_debug = LlamaDebugHandler(print_trace_on_end=True)
callback_manager = CallbackManager([llama_debug])
service_context = ServiceContext.from_defaults(
    callback_manager=callback_manager,
    llm=llm
)
```

```python
sql_database = SQLDatabase(engine)
```

```python
from llama_index.indices.struct_store.sql_query import NLSQLTableQueryEngine

sql_query_engine = NLSQLTableQueryEngine(
    sql_database=sql_database,
    service_context=service_context
)
```

```python
query_str = "What is the revenue of AMZN over time? Just generate the query."
response = sql_query_engine.query(query_str)

print(response.metadata['sql_query'])
print('=======')
print(response)
```

```python
with engine.connect() as conn:
    df = pd.read_sql(
        text(response.metadata['sql_query']),
        conn
    )
df
```

```python
sql_query_engine_tools = [
    QueryEngineTool(
        query_engine=sql_query_engine,
        metadata=ToolMetadata(
            name=f"sql_tool",
            description="""
            Useful for translating natural language query into a SQL query over a database containing financial information about TSLA.
            """
        )
    )
]
```

```python
query_engine = SubQuestionQueryEngine.from_defaults(
    query_engine_tools=sql_query_engine_tools,
    service_context=service_context
)

query_engine_tools = [
    QueryEngineTool(
        query_engine=query_engine,
        metadata=ToolMetadata(
            name="sub_question_query_engine",
            description=f"Useful when answering queries about {TICKER}",
        )
    )
]
```

```python
response = query_engine.query("Is AMZN running at a loss? Has this profit been increasing over time? Go step-by-step. And refer to the numbers.")
print("==============")
print(response)
```

```python
response = sql_query_engine.query("How much total debt does AMZN have at present? How much of it is short-term and how much is long-term?")
print(response)
```

```python
with engine.connect() as conn:
    df = pd.read_sql(text("select date,long_term_debt, short_term_debt from balance"), conn)

df
```

```python
tools = query_engine_tools + sql_query_engine_tools
tools
```

```python
# Create the agent
from llama_index.agent import OpenAIAgent, ReActAgent
from llama_index.llms import OpenAI
```

```python
agent = ReActAgent.from_tools(tools, verbose=True, llm=llm, max_iterations=20)
```

```python
agent.reset()
```

```python
response = agent.chat(
    """
    Do an evaluation of AMZN from a fundamentals point of view for the most recent year of data.
    Consider the earnings, income statements, cash, ratios and anything else you find useful that you can look-up from the data.
    Give your answers briefly in a bulleted list, refer to the numbers and explain your reasoning.
    Correct your errors, if you find any, with subsequent data lookups.
    Go step-by-step.
    """
)
print('============')
print(response)
```
